# Data Exploration
Explore the datasets used for threat detection across all 4 domains.

```python
import sys
sys.path.insert(0, '..')

from src.data.loader import DataLoader
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')
loader = DataLoader()
```

```python
# Load all domains
domains = ['phishing', 'network_intrusion', 'malware', 'log_analysis']
datasets = {}

for domain in domains:
    data = loader.load(domain)
    datasets[domain] = data
    n_malicious = sum(1 for d in data if d['label'] == 1)
    n_benign = sum(1 for d in data if d['label'] == 0)
    print(f'{domain}: {len(data)} samples ({n_malicious} malicious, {n_benign} benign)')
```

```python
# Distribution of samples per domain
fig, ax = plt.subplots(figsize=(10, 5))
domain_counts = {d: {'malicious': sum(1 for s in data if s['label']==1),
                      'benign': sum(1 for s in data if s['label']==0)}
                 for d, data in datasets.items()}

df_counts = pd.DataFrame(domain_counts).T
df_counts.plot(kind='bar', ax=ax, color=['#e74c3c', '#27ae60'])
ax.set_title('Sample Distribution by Domain', fontsize=14)
ax.set_ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Sample preview from each domain
for domain, data in datasets.items():
    print(f'\n{"="*50}')
    print(f'DOMAIN: {domain.upper()}')
    print(f'{"="*50}')
    if data:
        sample = data[0]
        print(f'Label: {"Malicious" if sample["label"] == 1 else "Benign"}')
        print(f'Text preview:\n{sample["text"][:500]}...')
```
